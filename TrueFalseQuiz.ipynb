```python
answers=['F','T','T','T','F','T','F','F','T','T']   
total_score=0
def CheckAns():
    QUIZTEXT.delete("1.0",END)
    global total_score
    global answers
    if(a1.get().upper()=="F"):
        total_score+=1
    if(a2.get().upper()=="T"):
        total_score+=1
    if(a3.get().upper()=="T"):
        total_score+=1
    if(a4.get().upper()=="T"):
        total_score+=1
    if(a5.get().upper()=="F"):
        total_score+=1
    if(a6.get().upper()=="T"):
        total_score+=1
    if(a7.get().upper()=="F"):
        total_score+=1
    if(a8.get().upper()=="F"):
        total_score+=1
    if(a9.get().upper()=="T"):
        total_score+=1
    if(a10.get().upper()=="T"):
        total_score+=1
    a1.set("")
    a2.set("")
    a3.set("")
    a4.set("")
    a5.set("")
    a6.set("")
    a7.set("")
    a8.set("")
    a9.set("")
    a10.set("")
    QUIZTEXT.insert(END,"\n\tCongratulations "+name.get()+" !")
    QUIZTEXT.insert(END,"\n\n\tYour total score is: "+str(total_score))
    QUIZTEXT.insert(END,"\n\n\tACCURACY LEVEL: "+str((total_score/10)*100))
    QUIZTEXT.insert(END,"\n\n\t THANK YOU! HOPE YOU LIKE IT!")
```
